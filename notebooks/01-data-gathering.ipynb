### 7 - Observations

- **No missing values** across all 205 rows.
- **Target (price):** right-skewed. Mean ($13,277) is well above median ($10,295), range $5,118 to $45,400. Consider log-transforming the target during preprocessing.
- **`car_ID`:** row index only (1-205). Drop it.
- **`CarName`:** 147 unique values, high cardinality. Contains typos (e.g. "alfa-romero" vs "alfa romeo") and inconsistent casing. Extract brand/make as a new feature.
- **`symboling`:** insurance risk rating (-2 to 3). Integer but ordinal/categorical in nature.
- **`doornumber` and `cylindernumber`:** stored as text ("two", "four") instead of numeric. Convert during cleaning.
- **`enginelocation`:** 202 of 205 values are "front", near-zero variance. Candidate for removal.
- **Numeric columns:** no obvious garbage values from `describe()`; ranges look plausible for car specifications.
